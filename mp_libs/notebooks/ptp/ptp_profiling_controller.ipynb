```python
# %%micropython

from machine import Pin
from time import sleep, sleep_ms

# Create a output pin, normally low
pin = Pin(1, mode=Pin.OUT, value=0)
sleep(30)

# Periodically, set pin high for 1 msec and then back low
while True:
    print("BOOM")
    pin.value(1)
    sleep_ms(10)
    pin.value(0)

    sleep(15)
```
